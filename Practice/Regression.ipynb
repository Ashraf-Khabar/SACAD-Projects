```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X_train, Y_train)
```

```python
plt.scatter(X, Y, color='blue', label='Données réelles')
plt.plot(X, linear_reg.predict(X), color='red', label='Régression linéaire')
```

```python
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, Y_train)
```

```python
plt.scatter(X, Y, color='blue', label='Données réelles')
plt.plot(X, tree_reg.predict(X), color='green', label='Decision Tree Regressor')

plt.title('Régression Linéaire et Autres Algorithmes')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

```python
forest_reg = RandomForestRegressor()
forest_reg.fit(X_train, Y_train)
```

```python
plt.scatter(X, Y, color='blue', label='Données réelles')
plt.plot(X, forest_reg.predict(X), color='orange', label='Random Forest Regressor')

plt.title('Régression Linéaire et Autres Algorithmes')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```
